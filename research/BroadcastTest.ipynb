```python
import numpy as np
import pandas as pd

a = np.arange(100, dtype=np.float64).reshape(10,10)
rep_count = 1000000
z = np.empty(rep_count)
print(a)

def work_func(*args):
    return args[0]+args[1]+args[2]

def with_tiles():
    a_tile = np.tile(a, (rep_count, 1))
    res = pd.DataFrame(work_func(*a_tile.T))
    print('tiles', res)
def with_bcast():
    _, a_broadcast = np.broadcast_arrays(z[:, np.newaxis, np.newaxis], a.T[np.newaxis, ...])
    #a_broadcast = a_broadcast.reshape(a.shape[0]*rep_count, a.shape[1])
    res = pd.DataFrame(work_func(*a_broadcast.transpose(1, 0, 2)))
    print('bcast', res)
```

```python
%memit with_tiles()
%memit with_bcast()
```
