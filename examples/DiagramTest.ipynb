```python
import calphad.libcalphadcpp as lcp
from calphad.minimize.globmin import GlobalMinimizer
import numpy as np
from scipy.spatial import ConvexHull
from scipy.spatial import distance_matrix
import calphad.plot.projections.triangular
import matplotlib.pyplot as plt
# Load thermodynamic database
maindb = lcp.Database("bigrose.tdb")
if maindb == None:
	sys.exit("Failed to load database")

test_elements = ['H','HE','LI','BE','B','C','N','O','F','NE']
# Generate conditions for 1-10 component systems
all_conditions = []
added_elements = []
for element in test_elements:
     added_elements.append(element)
     conds = lcp.evalconditions()
     conds.statevars['T'] = 2500
     conds.statevars['P'] = 101325
     conds.statevars['N'] = 1
     for added_element in added_elements:
          conds.elements.append(added_element)
     conds.phases["TEST"] = lcp.PhaseStatus.ENTERED
     all_conditions.append(conds)
```

```python
index = 2 # 3 component systemn
# Configure variable names
indices = lcp.IndexBiMap()
varmap = lcp.build_variable_map ( maindb.get_phases(), all_conditions[index], indices )
cmps = {}
# Create phases
for ph in maindb.get_phases():
    status = conds.phases[ph.key()]
    if (status == lcp.PhaseStatus.ENTERED):
        cmps[ph.key()] = lcp.CompositionSet(ph.data(), maindb.get_parameter_set(), varmap, indices)
# Initialize minimizer and sample points
engine = lcp.GlobalMinimizer()
engine.initial_subdivisions_per_axis = 150
engine.discard_unstable = False
points = engine.point_sample(cmps['TEST'], varmap, all_conditions[index])
```

```python
hull_vertices_collection = hull.points[hull.simplices][:,:,[0,1]]

from scipy.misc import factorial
def simplex_volume(coords):
    vertex_matrix = np.ones([coords.shape[0], coords.shape[0]])
    vertex_matrix[:,1:] = coords
    return np.linalg.det(vertex_matrix) / factorial(coords.shape[0]-1)

simplex_volumes = np.array([simplex_volume(simplex) for simplex in hull_vertices_collection])
print(simplex_volumes)
```

```python
import scipy.spatial.distance

point_mask = np.ones(len(hull.points), dtype=bool) # mask all points by default

# unmask any point that is an endpoint for a tieline
minimum_distance = (3 * np.sqrt(2)) / engine.initial_subdivisions_per_axis
simplex_mask = []
for simplex, equation, hull_vertices, volume in zip(hull.simplices, hull.equations, hull_vertices_collection, simplex_volumes):
    if np.any(np.any(np.where(points[simplex,0:2] < 1e-6, True, False),axis=1),axis=0): # this is a simplex of the upper hull; discard it
        continue
    distances = scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(hull_vertices))
    # figure out how to get the indices corresponding to the pairwise distance
    # use those indices to get the corresponding point indices in hull.simplices
    # use the point indices to unmask the correct points
    #print(np.where(distances > 0.05)[0]) # handle symmetry by only taking first element
    #print(simplex[np.where(distances > 0.05)[0]])
    point_mask[simplex[np.where(distances > minimum_distance)[0]]] = False # unmask these points

for simplex in hull.simplices:
    if np.all(~point_mask[simplex]): # if all vertices of this simplex are unmasked
        simplex_mask.append(False) # unmask this simplex
    else:
        simplex_mask.append(True)
   
#point_mask[:] = True # remask all
#point_mask[hull.simplices[simplex_mask].flatten()] = False # unmask points on unmasked simplices
```
