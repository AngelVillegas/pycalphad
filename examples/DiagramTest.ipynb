```python
import calphad.libcalphadcpp as lcp
from calphad.minimize.globmin import GlobalMinimizer
import numpy as np
from scipy.spatial import ConvexHull
from scipy.spatial import distance_matrix
import calphad.plot.projections.triangular
import matplotlib.pyplot as plt
# Load thermodynamic database
maindb = lcp.Database("bigrose.tdb")
if maindb == None:
	sys.exit("Failed to load database")

test_elements = ['H','HE','LI','BE','B','C','N','O','F','NE']
# Generate conditions for 1-10 component systems
all_conditions = []
added_elements = []
for element in test_elements:
     added_elements.append(element)
     conds = lcp.evalconditions()
     conds.statevars['T'] = 2700
     conds.statevars['P'] = 101325
     conds.statevars['N'] = 1
     for added_element in added_elements:
          conds.elements.append(added_element)
     conds.phases["TEST"] = lcp.PhaseStatus.ENTERED
     all_conditions.append(conds)
```

```python
index = 2 # 3 component systemn
# Configure variable names
indices = lcp.IndexBiMap()
varmap = lcp.build_variable_map ( maindb.get_phases(), all_conditions[index], indices )
cmps = {}
# Create phases
for ph in maindb.get_phases():
    status = conds.phases[ph.key()]
    if (status == lcp.PhaseStatus.ENTERED):
        cmps[ph.key()] = lcp.CompositionSet(ph.data(), maindb.get_parameter_set(), varmap, indices)
# Initialize minimizer and sample points
engine = lcp.GlobalMinimizer()
engine.initial_subdivisions_per_axis = 30
engine.refinement_subdivisions_per_axis = 5
engine.discard_unstable = True
points = engine.point_sample(cmps['TEST'], varmap, all_conditions[index])
```

```python
points = np.array(points)
```

```python
coord_indices = np.arange(index+2)
coord_indices = np.delete(coord_indices,[index],None) # TODO: Use automated deletion of dependent dimensions
if (len(points) > 3):
    hull = ConvexHull(points[:,coord_indices])
```

```python
import scipy.spatial.distance
import itertools

point_mask = np.ones(len(points), dtype=bool) # mask all points by default

# unmask any point that is an endpoint for a tieline
minimum_distance = (1 * np.sqrt(2)) / engine.initial_subdivisions_per_axis
simplex_mask = []
tie_lines = []
tie_line_colors = []
tie_line_widths = []
for simplex in hull.simplices:
    if np.any(np.any(np.where(points[simplex,0:2] < 1e-6, True, False),axis=1),axis=0): # this simplex is not part of the energy surface
        continue
    distances = scipy.spatial.distance.pdist(points[simplex,0:2])
    shortest_distance = np.min(distances)
    simplex_edges = np.asarray(list(itertools.combinations(simplex,2)))
    new_lines = simplex_edges[np.where(distances >= minimum_distance)]
    if len(simplex_edges) == len(new_lines):
        # This is a three phase region because all simplices satisfy the minimum distance requirement
        tie_lines.extend(new_lines)
        point_mask[new_lines] = False
        # Red for a tie plane
        tie_line_colors.append([1,0,0,1])
        tie_line_colors.append([1,0,0,1])
        tie_line_colors.append([1,0,0,1])
        tie_line_widths.append(3)
        tie_line_widths.append(3)
        tie_line_widths.append(3)
    elif len(new_lines) == 2:
        # This is a two phase region
        shortest_edge = simplex_edges[np.where(distances == shortest_distance)][0]
        other_tiepoint_index = np.setdiff1d(simplex, shortest_edge, assume_unique=True)[0]
        # For the short edge, we need to choose the vertex with the minimum energy
        # It is more likely to be on the true tie-line since nearby metastable points will be higher in energy
        minimum_energy_vertex = shortest_edge[np.argmax(points[shortest_edge][:,-1])]
        tie_lines.append([other_tiepoint_index, minimum_energy_vertex])
        point_mask[[other_tiepoint_index, minimum_energy_vertex]] = False
        # Green for a tie line
        tie_line_colors.append([0,1,0,1])
        tie_line_widths.append(0.5)
    elif len(new_lines) == 0:
        # Single-phase region; drop this simplex
        pass
tie_lines = np.asarray(tie_lines)
```
