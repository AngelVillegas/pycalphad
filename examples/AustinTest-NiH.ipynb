```python
# Initialize minimizer and sample points
engine = lcp.GlobalMinimizer()
engine.initial_subdivisions_per_axis = 20
engine.discard_unstable = False
points_1073 = engine.point_sample(cmps['FCC_A1'], varmap, conds)
conds.statevars['T'] = 1473
points_1473 = engine.point_sample(cmps['FCC_A1'], varmap, conds)
points_1073 = np.array(points_1073)
points_1473 = np.array(points_1473)
```
