```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import collections as mc
from sympy.printing.theanocode import theano_function
import theano
import sympy
import numpy as np
import pandas as pd
from scipy.spatial import ConvexHull
import pycalphad.minimize as cp
import pycalphad.plot.projections.triangular
from pycalphad.io.database import Database
from pycalphad.io.tdb import tdbread
from pycalphad.model import Model
from pycalphad.variables import *
```

```python
# monkey patch for theano_function handling sympy
sympy.printing.theanocode.mapping[sympy.And] = theano.tensor.and_
def _special_print_Piecewise(self, expr, **kwargs):
        import numpy as np
        from theano.ifelse import ifelse
        e, cond = expr.args[0].args
        if len(expr.args) == 1:
            return ifelse(self._print(cond, **kwargs),
                             self._print(e, **kwargs),
                             np.nan)
        return ifelse(self._print(cond, **kwargs),
                         self._print(e, **kwargs),
                         self._print(sympy.Piecewise(*expr.args[1:]), **kwargs))
sympy.printing.theanocode._print_Piecewise = _special_print_Piecewise
```

```python
db = Database()
tdbread(db, open('bigrose.tdb').read())
cur_phase = 'TEST'
temp = 2500 # Kelvin
mod = Model(db, ['H','HE','LI'], [cur_phase])
sympy_model = mod({T:temp})
# get all of the variables that haven't been filled in
test_vars = sympy_model.atoms(StateVariable)
# compile G for Theano
gibbs_energy = theano_function(test_vars, [sympy_model])
```

```python
subls = [3]
points = cp.point_sample(subls, size=10000)
gibbs_matrix = np.ndarray((len(points),sum(subls)+1))
energies = []
# TODO: not very efficient
for p in points:
    energies.append(gibbs_energy(*p))
energies = np.array(energies)
gibbs_matrix[:,0:-1] = points
gibbs_matrix[:,-1] = energies

# TODO: needs automated dependent dimension deletion
hull = ConvexHull(gibbs_matrix[:,[0,1,3]])
data_dict = {
'Phase' : cur_phase,
'GM' : energies,
'T' : temp,
'X1' : points[:,0],
'X2' : points[:,1]
}
hull_frame = pd.DataFrame(data_dict)
x_variable = 'X1'
y_variable = 'X2'
```

```python
scaled_energies = (energies - energies.min()) / energies.ptp()
colors = plt.cm.afmhot(scaled_energies)
fig = plt.figure(dpi=600,figsize=(9,9))
ax = fig.gca(projection="triangular") # use ternary axes
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(True)
plt.xlim([-0.01,1])
plt.ylim([-0.01,1])
plt.gca().set_aspect('equal')
plt.scatter(points[:,0], points[:,1], alpha=1, c=colors)
plt.show()
```

```python
import scipy.spatial
import itertools
point_mask = np.ones(len(energies), dtype=bool) # mask all points by default

# unmask any point that is an endpoint for a tieline
minimum_distance = 0.06
tie_lines = []
tie_line_colors = []
tie_line_widths = []
for simplex, eq in zip(hull.simplices, hull.equations):
    if eq[-2] > -1e-5:
        # simplex oriented 'upwards' in energy direction
        # must not be part of the energy surface
        continue
    if np.any(np.any(np.where(hull_frame[[x_variable, y_variable]].iloc[simplex] < 1e-3, True, False),axis=1),axis=0): 
        # this simplex is not part of the energy surface
        continue
    if np.all(np.any(np.where(hull_frame[[x_variable, y_variable]].iloc[simplex] > 0.98, True, False),axis=1),axis=0):
        # this simplex is not part of the energy surface
        continue
    distances = scipy.spatial.distance.pdist(hull_frame[[x_variable, y_variable]].iloc[simplex])
    shortest_distance = np.min(distances)
    simplex_edges = np.asarray(list(itertools.combinations(simplex,2)))
    phase_edge_list = hull_frame['Phase'].values[simplex_edges]
    edge_phases_match = np.array(phase_edge_list[:,0] == phase_edge_list[:,1])
    new_lines = simplex_edges[(distances >= minimum_distance) | (~edge_phases_match)] # tie line because of length
    # or tie line because it connects two different phases
    if len(new_lines) == 3:
        # This is a three phase region because all simplices satisfy the minimum distance requirement
        tie_lines.extend(new_lines)
        point_mask[new_lines] = False
        # Red for a tie plane
        tie_line_colors.append([1,0,0,1])
        tie_line_colors.append([1,0,0,1])
        tie_line_colors.append([1,0,0,1])
        tie_line_widths.append(3)
        tie_line_widths.append(3)
        tie_line_widths.append(3)
    elif len(new_lines) == 2:
        # This is a two phase region
        shortest_edge = simplex_edges[np.where(distances == shortest_distance)][0]
        other_tiepoint_index = np.setdiff1d(simplex, shortest_edge, assume_unique=True)[0]
        # For the short edge, we need to choose the vertex with the minimum energy
        # It is more likely to be on the true tie-line since nearby metastable points will be higher in energy
        minimum_energy_vertex = np.argmin(hull_frame['GM'].iloc[shortest_edge])
        tie_lines.append([other_tiepoint_index, minimum_energy_vertex])
        point_mask[[other_tiepoint_index, minimum_energy_vertex]] = False
        # Green for a tie line
        tie_line_colors.append([0,1,0,1])
        tie_line_widths.append(0.5)
    elif len(new_lines) == 0:
        # Single-phase region; drop this simplex
        pass
tie_lines = np.asarray(tie_lines)
```

```python
# Final plotting setup

x = hull_frame[x_variable]
y = hull_frame[y_variable]

# Mask the metastable points
masked_x = np.ma.array(hull_frame[x_variable], mask=point_mask)
masked_y = np.ma.array(hull_frame[y_variable], mask=point_mask)

fig = plt.figure(dpi=600,figsize=(12,12))
ax = fig.gca(projection="triangular") # use ternary axes
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(True)
plt.xlim([-0.01,1])
plt.ylim([-0.01,1])
plt.gca().set_aspect('equal')
lc = mc.LineCollection(np.asarray(hull_frame[[x_variable,y_variable]])[tie_lines], color=tie_line_colors, linewidth=tie_line_widths)
ax.add_collection(lc)
ax.scatter(masked_x, masked_y, color='black')

ax.text(0.3, 0.8, 'T = '+str(temp)+ ' K',
        verticalalignment='bottom', horizontalalignment='left',
        color='black', fontsize=20)


plt.title('Rose Diagram',fontsize=25, x=0.5, y=0.9)
ax.set_xlabel(x_variable, labelpad=15,fontsize=20)
ax.set_ylabel(y_variable,rotation=60,fontsize=20,labelpad=-120)
plt.show()
```
