```python
import calphad.libcalphadcpp as lcp
from calphad.minimize.globmin import GlobalMinimizer
import calphad.plot.projections.triangular
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load thermodynamic database
maindb = lcp.Database("crfeni_mie.tdb")
if maindb == None:
	sys.exit("Failed to load database")

# Set equilibrium conditions
conds = lcp.evalconditions()
conds.statevars['T'] = 1650
conds.statevars['P'] = 101325
conds.statevars['N'] = 1
conds.elements.append("FE")
conds.elements.append("NI")
conds.elements.append("CR")
conds.elements.append("VA")
conds.xfrac["NI"] = .08
conds.xfrac["CR"] = .18
conds.phases["HCP_A3"] = lcp.PhaseStatus.SUSPENDED
conds.phases["BCC_A2"] = lcp.PhaseStatus.ENTERED
conds.phases["FCC_A1"] = lcp.PhaseStatus.ENTERED
conds.phases["LIQUID"] = lcp.PhaseStatus.ENTERED
conds.phases["SIGMA"] = lcp.PhaseStatus.SUSPENDED

indices = lcp.IndexBiMap()
varmap = lcp.build_variable_map ( maindb.get_phases(), conds, indices )
cmps = {}

for ph in maindb.get_phases():
    status = conds.phases[ph.key()]
    if (status == lcp.PhaseStatus.ENTERED):
        cmps[ph.key()] = lcp.CompositionSet(ph.data(), maindb.get_parameter_set(), varmap, indices)
```

```python
%time globminengine = GlobalMinimizer(cmps, varmap, conds, initial_subdivisions_per_axis=30)
```

```python
%time hull_frame, triangles = globminengine.triangulate_hull()
```

```python
# Perform some setup for phase-specific coloring of points

colorlist = {}

# colors from Junwei Huang, March 21 2013
ColorValues=["FF0000", "00FF00", "0000FF", "FFFF00", "FF00FF", "00FFFF", "000000", 
        "800000", "008000", "000080", "808000", "800080", "008080", "808080", 
        "C00000", "00C000", "0000C0", "C0C000", "C000C0", "00C0C0", "C0C0C0", 
        "400000", "004000", "000040", "404000", "400040", "004040", "404040", 
        "200000", "002000", "000020", "202000", "200020", "002020", "202020", 
        "600000", "006000", "000060", "606000", "600060", "006060", "606060", 
        "A00000", "00A000", "0000A0", "A0A000", "A000A0", "00A0A0", "A0A0A0", 
        "E00000", "00E000", "0000E0", "E0E000", "E000E0", "00E0E0", "E0E0E0"]
M=len(ColorValues)

phasecount = 0
for phase in conds.phases:
	colorlist[phase.key()] = "#"+ColorValues[np.mod(phasecount,M)]
	phasecount = phasecount + 1

# Get the configured plot colors
plotcolors = map(lambda x: colorlist[x], hull_frame['Phase'])
```

```python
# Final plotting setup

x_variable = 'X(CR)'
y_variable = 'X(NI)'
x = hull_frame[x_variable]
y = hull_frame[y_variable]
triangles = np.asarray(triangles)

# Mask the metastable points
masked_x = np.ma.masked_array(x, mask=~hull_frame['Stable'])
masked_y = np.ma.masked_array(y, mask=~hull_frame['Stable'])

fig = plt.figure(dpi=600,figsize=(9,9))
ax = fig.gca(projection="triangular") # use ternary axes
ax.tick_params(axis='both', which='major', labelsize=14)
ax.grid(True)
plt.xlim([-0.01,1])
plt.ylim([-0.01,1])
plt.gca().set_aspect('equal')
plt.triplot(masked_x, masked_y, triangles, 'go-', color='k', marker=None, zorder=1)

ax.text(0.3, 0.8, 'T = '+str(conds.statevars['T'])+ ' K',
        verticalalignment='bottom', horizontalalignment='left',
        color='black', fontsize=20)

for phase in conds.phases:
    # Plot each phase separately so we can mark the legend properly
    phase_frame = hull_frame[(hull_frame.Stable == True) & (hull_frame.Phase == phase.key())]
    if (len(phase_frame) > 0):
        ax.scatter(phase_frame[x_variable], phase_frame[y_variable], c=colorlist[phase.key()],
                       marker='o',
                       label=phase.key(),
                       zorder=2)
        ax.legend(loc=0, fontsize=15)
plt.title('Using DataFrames',fontsize=25, x=0.5, y=0.9)
ax.set_xlabel(x_variable, labelpad=15,fontsize=20)
ax.set_ylabel(y_variable,rotation=60,fontsize=20,labelpad=-120)
plt.show()
```
